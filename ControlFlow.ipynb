- A program’s control flow is in which the program’s code executes.
- The control flow of a python program is regulated by conditional statements, loops, and functional calls.
- Three types of Control structures:
-- Sequential: default mode
-- Selection: used for decision and branching
-- Repetition: used for Looping
<br>
<br>
1. Sequential: Sequential statements are a set of statements whose execution process happens in a sequence. The problem with sequential statements is that if the logic has broken in any one of the lines, then the complete source code execution will break.
2. Selection/Decision control statements: In Python, the selection statements are also known as Decision control statements or branching statements.
The selection statement allows a program to test several conditions and execute instructions based on which condition is true.
Some Decision Control Statements are:
- Simple if
- if-else
- nested if
- if-elif-else

Condition.png

- Simple if: If statements are control flow statements that help us to run a particular code, but only when a certain condition is met or satisfied A simple if only has one condition to check.
- If-else: The if-else Statement  Evaluates the condition and will execute the body  of if elf the test condition is False , then body of else is executed.
- Nested if: Nested if statements are an if statement inside another if statement.
- if-elif-else: The if-elif-else statement is used to conditionally execute a statement or a block of statements.

```python
x,y=6,10
if x>y:
    print(x)
else:
    print(y)
```

```python
n=int(input())

if(n%2==0):
    print("EVEN")
else:
    print("ODD")
```

```python
#even number and divisible by 3
n=int(input())

if n%2==0:
    print("This is a even number")
    if n%3==0:
        print("even number divisible by 3")
    else:
        print("even number not divisible by 3")

elif n%3==0:
    print("Its divisible by 3")
    if n%5==0:
        print("Number is divisible by 3 and 5")
    else:
        print("Number is divisible by 3 but not by 5")
else:
    print("This is a random number")
```
