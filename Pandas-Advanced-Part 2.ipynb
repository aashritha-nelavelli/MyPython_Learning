```python
import pandas as pd
```

```python
df_diabetes=pd.read_csv('diabetes.csv')
df_diabetes.head()
```

```python
def replace_val(x):
    if x==1:
        return "Has Diabetes"
    else:
        return "No Diabetes"
```

```python
df_diabetes['Outcome_by_Text']=df_diabetes["Outcome"].apply(replace_val)
df_diabetes.head()
```

```python
df_diabetes['sum'] = df_diabetes.apply(lambda x: x['Age'] + x['BMI'], axis=1)
df_diabetes.head()
```

```python
df_diabetes['diff'] = df_diabetes.apply(lambda x: x['Age'] - x['BMI'], axis=1)
df_diabetes.head()
```

```python
df_diabetes['prod'] = df_diabetes.apply(lambda x: x['Age'] * x['BMI'], axis=1)
df_diabetes.head()
```

```python
df_diabetes['quotent'] = df_diabetes.apply(lambda x: x['BMI']/x['Age'], axis=1)
df_diabetes.head()
```
