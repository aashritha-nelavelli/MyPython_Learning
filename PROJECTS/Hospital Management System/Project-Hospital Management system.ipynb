```python
import pandas as pd
```

```python
df_hospital_data=pd.read_csv('Hospital_Management.csv')
df_hospital_data
```

```python
df_doc_data=pd.read_csv("Doctor_entry.csv")
df_doc_data
```

```python
from datetime import date
from datetime import datetime

def patient_id_generation(name,date):
    return name.upper()[:3]+date[:2]+date[3:5]

def patient_entry():
    
    entryNumber=input("entry number")
    name=input("Enter Patient Name: ")
    patient_age=input("Age ")
    visit_date=date.today()
    dob=input("Enter Patient Date of Birth (DD-MM-YYYY): ")
    Patient_Address=input("Enter Address: ")
    Payment_Mode=input("Payment by ")
    Membership=input("Membership ")
    patient_id=patient_id_generation(name,dob)
    Token='RKC'+visit_date.strftime("%m%d%Y")+'00'+entryNumber
    
    
    df_temp=pd.DataFrame()

    df_temp['Patient_ID']=[patient_id]
    df_temp['Visit_Date']=[visit_date]
    df_temp['DOB']=[dob]
    df_temp['Age']=[patient_age]
    df_temp['Patient_Name']=[name]
    df_temp['Patient_Address']=[Patient_Address]
    df_temp['Payment_Mode']=[Payment_Mode]
    df_temp['Membership']=[Membership]
    df_temp['Token']=[Token]
    
    return df_temp

def doc_entry():
    
    df_temp=pd.DataFrame()
    Token=input("Token ")
    patient_id=df_hospital_data[df_hospital_data['Token']==Token]['Patient_ID'].tolist()[0]
    visit_date=date.today()
    Doctor_Comments=input("Comments: ")
    Remarks=input("Remarks ")
    Revisit_Required=input("Revisit Required ")
    
    Preciption_Number=patient_id+visit_date.strftime("%m%d%Y")
    
    df_temp['Patient_ID']=[patient_id]
    df_temp['Visit_Date']=[visit_date]
    df_temp['Token']=[Token]
    df_temp['Doctor_Comments']=[Doctor_Comments]
    df_temp['Remarks']=[Remarks]
    df_temp['Revisit_Required']=[Revisit_Required]
    df_temp['Preciption_Number']=[Preciption_Number]
    
    return df_temp
    
    
def payment_entry():
    df_temp=pd.DataFrame()
    Token=input("Token ")
    patient_id=df_hospital_data[df_hospital_data['Token']==Token]['Patient_ID'].tolist()[0]
    Payment_mode=input("Cash or Card")
    Membership=input("Yes or No")
    
    Preciption_Number=patient_id+visit_date.strftime("%m%d%Y")
    
    df_temp['Patient_ID']=[patient_id]
    df_temp['Payment_mode']=[Card or Cash]
    df_temp['Membership']=[Membership]
    df_temp['Token']=[Token]
    
    return df_temp
```

```python
df_hospital_data=df_hospital_data.append(patient_entry())
df_hospital_data
# df_hospital_data.to_csv("Hospital_Management.csv")
```

```python
df_doc_data=df_doc_data.append(doc_entry())
df_doc_data
# df_doc_data.to_csv("Doctor_entry.csv")
```
