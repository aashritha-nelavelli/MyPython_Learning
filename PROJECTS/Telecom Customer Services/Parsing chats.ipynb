```python
import pandas as pd
```

```python
file1 = open('Makingofdummydata.txt', 'r')
Lines = file1.readlines()
 
line_list=[]
# Strips the newline character
for line in Lines:
    line_list.append(line.strip())
```

```python
line_list
```

```python
texts_list=[]

for i in line_list:
    if "(2" in i:
        texts_list.append(i)
```

```python
texts_list
```

```python
chats=[]
time_stamp=[]
for i in texts_list:
    chats.append(i.split(")")[1])
    time_stamp.append(i.split(")")[0][1:])
```

```python
df=pd.DataFrame()
df["Time_Stamp"]=time_stamp
df["Chats"]=chats

df
```

```python
list1=df['Chats'].to_list()
chats=[]
caller=[]

for i in list1:
    chats.append(i.split(":")[1])
    caller.append(i.split(":")[0])
```

```python
df=pd.DataFrame()
df["Time_Stamp"]=time_stamp
df["Caller"]=caller
df["chats"]=chats


df
```

```python
df.to_excel("chats.xlsx")
```
